观察下方的分治算法（divide-and-conquer algorithm）的伪代码, 回答下面问题

DoSomething(A,p,r)

---

**2. 算法功能与效率分析**

该算法的功能是**对数组进行排序**。

分析：当 $n=2$ 时，若两元素逆序则交换，使两个元素有序。当 $n \ge 3$ 时，先对前 $2n/3$ 个元素排序，再对后 $2n/3$ 个元素排序，再对前 $2n/3$ 个元素排序。通过重叠子数组的反复排序，最终使整个数组有序。该算法是 Stooge Sort 的变体。

 该算法的时间复杂度为 $\Theta(n^{2.71})$，远高于基于比较的排序算法的最优下界 $O(n \log n)$。更高效的排序算法包括归并排序（Merge Sort）、快速排序（Quick Sort）、堆排序（Heap Sort），它们的时间复杂度均为 $O(n \log n)$。

---

**3. Python 实现**

````python
import math

def do_something(A, p, r):
    """原始分治排序算法（Stooge Sort 变体）"""
    n = r - p + 1
    if n == 2 and A[p] > A[r]:
        A[p], A[r] = A[r], A[p]
    elif n >= 3:
        m = math.ceil(2 * n / 3)
        do_something(A, p, p + m - 1)
        do_something(A, r - m + 1, r)
        do_something(A, p, p + m - 1)


def merge_sort(A):
    """更高效的归并排序 O(n log n)"""
    if len(A) <= 1:
        return A
    mid = len(A) // 2
    left = merge_sort(A[:mid])
    right = merge_sort(A[mid:])
    return merge(left, right)


def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result


# 测试原始算法
A1 = [5, 3, 8, 1, 2, 7, 4, 6]
do_something(A1, 0, len(A1) - 1)
print("do_something 排序结果:", A1)

# 测试高效算法
A2 = [5, 3, 8, 1, 2, 7, 4, 6]
print("merge_sort 排序结果:", merge_sort(A2))

# your algorithm time complexity is: 原始算法 O(n^{log_{1.5} 3}) ≈ O(n^{2.71})；归并排序 O(n log n)
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

**解答：**

**算法设计思路：**

这是一个经典的"最大子数组和"问题的变体。我们需要找到 $i < j$ 使得利润最大。第 $i$ 天买入、第 $j$ 天卖出的利润 = profits[i+1] + ... + profits[j]，即求 profits 数组的最大子数组和。

**Kadane 算法思路：**
- 遍历数组，维护当前累计收益 `current_sum`
- 如果 `current_sum` 变为负数，则重置为0，并更新买入日期为下一天
- 在每一步，如果 `current_sum` 超过历史最大收益，则更新最大收益和卖出日期

**时间复杂度：$O(n)$**，只需一次遍历。
**空间复杂度：$O(1)$**。

部分算法思路和代码借鉴大模型辅助。

````python
def max_profit(profits):
    """
    计算最大利润及买入卖出日期
    返回: (max_profit, buy_day, sell_day)，日期从1开始计数
    """
    n = len(profits)
    if n < 2:
        return 0, -1, -1
    
    max_sum = float("-inf")
    current_sum = 0
    buy = 0          # 当前子数组的起始索引（买入日）
    best_buy = 0
    best_sell = 0
    
    for i in range(n):
        current_sum += profits[i]
        
        if current_sum > max_sum:
            max_sum = current_sum
            best_buy = buy
            best_sell = i
        
        if current_sum < 0:
            current_sum = 0
            buy = i + 1  # 下一个可能的买入日
    
    # 日期从1开始
    return max_sum, best_buy + 1, best_sell + 1


# 测试
profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
profit, buy_day, sell_day = max_profit(profits)
print(f"最大利润: {profit}, 买入日期: 第{buy_day}天, 卖出日期: 第{sell_day}天")

# your algorithm time complexity is: O(n)
````

## 问题 4

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

**解答：**

**1. $O(\log n)$ 找一个局部最小值 — 算法设计思路：**

利用二分查找思想：
- 检查中间元素 `A[mid]`
- 如果 `A[mid]` 比左右邻居都小，则找到局部最小值
- 否则，如果 `A[mid-1] < A[mid]`，则左侧存在局部最小值（因 $A[1] > A[2]$ 保证左端呈下降趋势）
- 否则，右侧存在局部最小值（因 $A[n-1] < A[n]$ 保证右端呈上升趋势）

**正确性：** 已知 $A[1] > A[2]$（左边界下降），$A[n-1] < A[n]$（右边界上升）。若 $A[mid-1] < A[mid]$，则在区间 $[1, mid]$ 内，两端均呈下降趋势向中间，由连续性可知必存在局部最小值。同理可证另一侧。

**时间复杂度：** 每次将搜索区间减半，$T(n) = T(n/2) + O(1)$，故 $T(n) = O(\log n)$。

---

**2. 找出所有局部最小值 — 算法设计思路：**

遍历数组，依次检查每个内部元素（索引 $2$ 到 $n-1$）是否比左右邻居都小。边界元素 $A[1]$ 和 $A[n]$ 不可能是局部最小值（因为题目条件 $A[1] > A[2]$ 且 $A[n-1] < A[n]$）。

**时间复杂度：$O(n)$**，需要遍历整个数组一次。这是最优的，因为要找出所有局部最小值，必须检查每个元素。

部分算法思路和代码借鉴大模型辅助。

````python
def find_one_local_min(A):
    """
    在 O(log n) 时间内找到一个局部最小值（返回数值）
    """
    n = len(A)
    left, right = 0, n - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        # 检查是否为局部最小
        left_smaller = (mid == 0) or (A[mid] < A[mid - 1])
        right_smaller = (mid == n - 1) or (A[mid] < A[mid + 1])
        
        if left_smaller and right_smaller:
            return A[mid]
        
        # 向左或向右搜索
        if mid > 0 and A[mid - 1] < A[mid]:
            right = mid - 1
        else:
            left = mid + 1
    
    return None


def find_all_local_mins(A):
    """
    在 O(n) 时间内找出所有局部最小值
    """
    n = len(A)
    result = []
    for i in range(1, n - 1):
        if A[i] < A[i - 1] and A[i] < A[i + 1]:
            result.append(A[i])
    return result


# 测试
A = [9, 3, 7, 2, 1, 4, 5]
print("找到一个局部最小值:", find_one_local_min(A))
print("所有局部最小值:", find_all_local_mins(A))

# your algorithm time complexity is: 找一个 O(log n)；找所有 O(n)
````

## 问题 5

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

**解答：**

**1. 排序法 — 归并排序**

选择归并排序（Merge Sort），时间复杂度稳定为 $O(n \log n)$。排序后直接取前 $k$ 个元素，总时间复杂度为 $O(n \log n)$。

---

**2. 堆方法**

通过逐个插入元素构建最小堆：每次插入的时间复杂度为 $O(\log i)$（其中 $i$ 为当前堆的大小），建堆总时间为：
$$\sum_{i=1}^{n} O(\log i) = O(n \log n)$$
然后调用 $k$ 次 Extract-min，每次 $O(\log n)$，共计 $O(k \log n)$。
总时间复杂度：$O(n \log n + k \log n)$。
注：若使用 Floyd 建堆法（自底向上），可将建堆优化到 $O(n)$。

---

**3. 海量数据 — 大小为k的最大堆**

由于数据总量超出内存，但可以存下 $k$ 个数据，因此使用一个大小为 $k$ 的最大堆（Max-Heap）：
- 遍历数据流中的每个元素
- 若堆中元素不足 $k$ 个，直接插入
- 若堆已满且当前元素小于堆顶（最大值），则弹出堆顶并插入当前元素
- 遍历完成后，堆中即为前 $k$ 小的数，排序输出

时间复杂度：每个元素处理时间为 $O(\log k)$，总共 $O(n \log k)$。
空间复杂度：$O(k)$。

部分算法思路和代码借鉴大模型辅助。

````python
import heapq

# ===== 方法1：排序法 O(n log n) =====
def k_smallest_sort(arr, k):
    """排序后取前k个"""
    sorted_arr = sorted(arr)  # Python 使用 Timsort，O(n log n)
    return sorted_arr[:k]


# ===== 方法2：堆插入法 O(n log n + k log n) =====
def k_smallest_heap_insert(arr, k):
    """逐个插入建堆，然后 k 次 extract-min"""
    heap = []
    for x in arr:
        heapq.heappush(heap, x)  # 逐个插入建堆 O(n log n)
    result = []
    for _ in range(k):
        result.append(heapq.heappop(heap))  # k 次 extract-min O(k log n)
    return result


# ===== 方法3：大小为k的最大堆（海量数据场景）O(n log k) =====
def k_smallest_large_data(arr, k):
    """
    使用大小为k的最大堆处理海量数据
    遍历每个元素，维护当前最小的k个元素
    """
    if k == 0:
        return []
    max_heap = []  # 用负数模拟最大堆（Python heapq 是最小堆）
    for x in arr:
        if len(max_heap) < k:
            heapq.heappush(max_heap, -x)
        elif -x > max_heap[0]:  # 即 x < -max_heap[0]（小于当前堆中最大值）
            heapq.heapreplace(max_heap, -x)
    result = sorted([-x for x in max_heap])
    return result


# 测试
arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print("方法1（排序）:", k_smallest_sort(arr, k))
print("方法2（插入建堆）:", k_smallest_heap_insert(arr, k))
print("方法3（k大小最大堆）:", k_smallest_large_data(arr, k))

# your algorithm time complexity is: 方法1 O(n log n)；方法2 O(n log n + k log n)；方法3 O(n log k)
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

**解答：**

**算法设计思路 — 栈模拟：**

使用栈来模拟星球碰撞过程。从左到右遍历每个星球：
- 正数（向右运动）直接入栈，因为其右边可能有向左运动的星球与之碰撞
- 负数（向左运动）可能与栈顶向右运动的星球碰撞：
  - 当栈顶为正数（向右）且 `|栈顶| < |当前星球|` 时：栈顶星球被摧毁（pop），继续检查
  - 当栈顶为正数且 `|栈顶| == |当前星球|` 时：质量相同，向左运动（当前星球）消失，向右运动（栈顶）保留，不push当前星球
  - 当栈顶为正数且 `|栈顶| > |当前星球|` 时：当前星球被摧毁，不push
  - 当栈为空或栈顶为负数（向左）时：当前星球直接入栈

**正确性分析：** 栈中始终保持的是在当前位置左侧已"存活"的星球序列。由于速度相同，只有向右的星球会与后面向左的星球碰撞。正数入栈等待未来可能的碰撞，负数则与栈中最近的正数碰撞。

**时间复杂度：$O(n)$**，每个星球至多入栈一次、出栈一次。
**空间复杂度：$O(n)$**，用于存储栈。

部分算法思路和代码借鉴大模型辅助。

````python
def planet_collision(A):
    """
    模拟星球碰撞，返回最终存活的星球数组
    """
    stack = []
    for planet in A:
        if planet > 0:
            # 向右运动，直接入栈
            stack.append(planet)
        else:
            # 向左运动，与栈顶向右运动的星球碰撞
            destroyed = False
            while stack and stack[-1] > 0:
                if abs(stack[-1]) < abs(planet):
                    # 栈顶质量小，被摧毁
                    stack.pop()
                elif abs(stack[-1]) == abs(planet):
                    # 质量相同，当前（向左）消失，栈顶（向右）保留
                    destroyed = True
                    break
                else:
                    # 栈顶质量大，当前星球被摧毁
                    destroyed = True
                    break
            if not destroyed:
                stack.append(planet)
    return stack


# 测试
print("测试1:", planet_collision([-3, -6, 2, 8, 5, -8, 9, -2, 1]))
print("测试2:", planet_collision([23, -8, 9, -3, -7, 9, -23, 22]))

# your algorithm time complexity is: O(n)
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

**解答：**

**算法设计思路 — 动态规划：**

定义 `dp[i][w]` 表示考虑前 $i$ 个物品、背包容量为 $w$ 时能获得的最大价值。

状态转移方程：
$$dp[i][w] = \max(dp[i-1][w], \ dp[i-1][w - w_i] + v_i) \quad \text{if } w \ge w_i$$
$$dp[i][w] = dp[i-1][w] \quad \text{if } w < w_i$$

边界条件：$dp[0][w] = 0$（没有物品时价值为0）。

可以使用滚动数组优化空间为 $O(W)$ 的一维DP：
$$dp[w] = \max(dp[w], \ dp[w - w_i] + v_i)$$
其中 $w$ 从 $W$ 向 $w_i$ 逆序遍历，确保每个物品只被使用一次。

**时间复杂度：$O(n \cdot W)$**，其中 $n$ 为物品数量，$W$ 为背包容量。
**空间复杂度：$O(W)$**（使用一维DP优化）。

对于本题：$n=4, W=9$。
最优选择：物品2 (4,40) + 物品4 (3,50)，总重 7，总价值 **90**。

部分算法思路和代码借鉴大模型辅助。

````python
def knapsack_01(items, capacity):
    """
    0-1 背包问题（一维DP优化）
    items: [(weight, value), ...]
    capacity: 背包最大承重
    返回: 最大价值
    """
    dp = [0] * (capacity + 1)
    for weight, value in items:
        for w in range(capacity, weight - 1, -1):
            dp[w] = max(dp[w], dp[w - weight] + value)
    return dp[capacity]


def knapsack_with_items(items, capacity):
    """返回最大价值和所选物品"""
    n = len(items)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        w_i, v_i = items[i - 1]
        for w in range(capacity + 1):
            if w >= w_i:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - w_i] + v_i)
            else:
                dp[i][w] = dp[i - 1][w]
    
    # 回溯找出所选物品
    selected = []
    w = capacity
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i - 1][w]:
            selected.append(items[i - 1])
            w -= items[i - 1][0]
    
    return dp[n][capacity], selected[::-1]


# 测试
items = [(5, 10), (4, 40), (6, 30), (3, 50)]
capacity = 9
max_val = knapsack_01(items, capacity)
print(f"最大价值: {max_val}")

max_val2, selected = knapsack_with_items(items, capacity)
print(f"最大价值: {max_val2}, 所选物品(重量,价值): {selected}")

# your algorithm time complexity is: O(n * W)，其中 n 为物品数量，W 为背包容量
````

## 问题 8

给你二叉搜索树的根节点 root ，该树中的恰好两个节点的值被错误地交换。请在不改变其结构的情况下，恢复这棵树 。设计算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g.:
---
1.

<div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

**解答：**

**算法设计思路 — 中序遍历：**

二叉搜索树（BST）的中序遍历结果应为一个严格递增序列。若恰好有两个节点被交换，则中序遍历中会出现逆序：
- 第一处：前一个元素 > 当前元素，记 `first = prev`（较大的那个被交换到了前面）
- 第二处：前一个元素 > 当前元素，记 `second = curr`（较小的那个被交换到了后面）

找到这两个节点后交换它们的值即可恢复BST。

特殊情况：若两个交换的节点在中序遍历中是相邻的，则只会检测到一处逆序，此时 `first = prev`，`second = curr`。

**时间复杂度：$O(n)$**，其中 $n$ 为节点数，只需一次中序遍历。
**空间复杂度：$O(h)$**，递归栈深度为树高，最坏 $O(n)$，平均 $O(\log n)$。
若使用 Morris 遍历，可达到 $O(1)$ 额外空间。

部分算法思路和代码借鉴大模型辅助。

````python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def recover_tree(root):
    """
    恢复被错误交换两个节点的BST
    """
    first = second = prev = None
    
    def inorder(node):
        nonlocal first, second, prev
        if not node:
            return
        inorder(node.left)
        
        if prev and prev.val > node.val:
            if not first:
                first = prev
            second = node
        prev = node
        
        inorder(node.right)
    
    inorder(root)
    
    if first and second:
        first.val, second.val = second.val, first.val


def build_tree(arr):
    """从 LeetCode 风格的列表构建二叉树"""
    if not arr:
        return None
    nodes = [TreeNode(x) if x is not None else None for x in arr]
    kids = nodes[::-1]
    root = kids.pop()
    for node in nodes:
        if node:
            if kids:
                node.left = kids.pop()
            if kids:
                node.right = kids.pop()
    return root


def tree_to_list(root):
    """将二叉树转换为 LeetCode 风格的列表"""
    if not root:
        return []
    result = []
    queue = [root]
    while queue:
        node = queue.pop(0)
        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)
    while result and result[-1] is None:
        result.pop()
    return result


# 测试1: [1,3,null,null,2] -> [3,1,null,null,2]
root1 = build_tree([1, 3, None, None, 2])
print("恢复前:", tree_to_list(root1))
recover_tree(root1)
print("恢复后:", tree_to_list(root1))

# 测试2: [3,1,4,null,null,2] -> [2,1,4,null,null,3]
root2 = build_tree([3, 1, 4, None, None, 2])
print("恢复前:", tree_to_list(root2))
recover_tree(root2)
print("恢复后:", tree_to_list(root2))

# your algorithm time complexity is: O(n)，其中 n 为节点数
````

## 问题 9 \*
给定 `n` 个非负整数表示每个宽度为 `1` 的柱子的高度图，计算按此排列的柱子，下雨之后能接多少雨水。



<div style="text-align:center;width:800px;"> <img alt="" src="./fig/hw3q6.png"> </div>

请分析算法设计思路，计算时间复杂度，并基于python编程实现

**解答：**

**算法设计思路 — 双指针法：**

对于每个位置 $i$，它能接住的雨水量 = $\min(left\_max, right\_max) - height[i]$，其中 $left\_max$ 是 $i$ 左侧最高柱子的高度，$right\_max$ 是 $i$ 右侧最高柱子的高度。

使用双指针优化空间：
- `left` 从左向右移动，`right` 从右向左移动
- 维护 `left_max`（左侧已遍历的最高柱子）和 `right_max`（右侧已遍历的最高柱子）
- 若 `left_max < right_max`，则 `left` 位置的雨水量由 `left_max` 决定，处理 `left` 并右移
- 否则，`right` 位置的雨水量由 `right_max` 决定，处理 `right` 并左移

**正确性：** 当 `left_max < right_max` 时，实际右最高值 $\ge right_max > left_max$，因此 $\min(左最高, 右最高) = left_max$，可以安全计算。

**时间复杂度：$O(n)$**，只需一次遍历。
**空间复杂度：$O(1)$**，只使用常数个变量。

部分算法思路和代码借鉴大模型辅助。

````python
def trap(height):
    """
    计算接雨水的总量
    双指针法：O(n) 时间，O(1) 空间
    """
    if not height:
        return 0
    
    left, right = 0, len(height) - 1
    left_max = right_max = 0
    water = 0
    
    while left < right:
        if height[left] < height[right]:
            if height[left] >= left_max:
                left_max = height[left]
            else:
                water += left_max - height[left]
            left += 1
        else:
            if height[right] >= right_max:
                right_max = height[right]
            else:
                water += right_max - height[right]
            right -= 1
    
    return water


# 测试
height1 = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
print(f"接雨水量: {trap(height1)}")  # 期望输出: 6

height2 = [4, 2, 0, 3, 2, 5]
print(f"接雨水量: {trap(height2)}")  # 期望输出: 9

# your algorithm time complexity is: O(n)
````

## 问题 10

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

> 例如，从根节点到叶节点的路径 1 -> 2 -> 3 表示数字 123 。

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

**解答：**

**算法设计思路 — DFS（深度优先搜索）：**

从根节点开始进行深度优先遍历，维护从根到当前节点的路径数值 `current_num`：
- 进入节点时：`current_num = current_num * 10 + node.val`
- 到达叶子节点（左右子节点均为空）时：将 `current_num` 累加到总和中
- 递归处理左右子树

**时间复杂度：$O(n)$**，其中 $n$ 为节点数，每个节点访问一次。
**空间复杂度：$O(h)$**，递归栈深度为树高，最坏 $O(n)$，平均 $O(\log n)$。

部分算法思路和代码借鉴大模型辅助。

````python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def sum_numbers(root):
    """
    计算从根到叶节点生成的所有数字之和
    """
    def dfs(node, current_num):
        if not node:
            return 0
        current_num = current_num * 10 + node.val
        # 叶子节点：返回当前路径数值
        if not node.left and not node.right:
            return current_num
        # 非叶子节点：递归求和
        return dfs(node.left, current_num) + dfs(node.right, current_num)
    
    return dfs(root, 0)


def build_tree(arr):
    """从 LeetCode 风格的列表构建二叉树"""
    if not arr:
        return None
    nodes = [TreeNode(x) if x is not None else None for x in arr]
    kids = nodes[::-1]
    root = kids.pop()
    for node in nodes:
        if node:
            if kids:
                node.left = kids.pop()
            if kids:
                node.right = kids.pop()
    return root


# 测试1: [1,2,3] -> 25
root1 = build_tree([1, 2, 3])
print(f"根到叶子数字之和: {sum_numbers(root1)}")  # 期望: 25

# 测试2: [4,9,0,5,1] -> 1026
root2 = build_tree([4, 9, 0, 5, 1])
print(f"根到叶子数字之和: {sum_numbers(root2)}")  # 期望: 1026

# your algorithm time complexity is: O(n)，其中 n 为节点数
````
